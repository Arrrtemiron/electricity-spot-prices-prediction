```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../datasets/atsenergo_2017_2023_1.csv')
data.head()
```

```python
data.index = data['timestamp']
data.index = pd.to_datetime(data.index)
data.drop(columns=['timestamp'], inplace=True)
data.price.plot()
plt.title('Prices from 2017 to 2023');
```

```python
data.resample("M").mean().plot()
plt.title(' ');
```
